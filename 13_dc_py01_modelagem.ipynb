Engenharia de features

```python
df_sisagua.drop(df_sisagua[["ano", "mes"]], axis=1, inplace=True)
df_sisagua['data_da_coleta'] = pd.to_datetime(df_sisagua['data_da_coleta']) # Convertendo o tipo do dado para datetime
df_sisagua['ano'] = df_sisagua['data_da_coleta'].dt.year # Extraindo o ano
df_sisagua['mes'] = df_sisagua['data_da_coleta'].dt.month # Extraindo o mês
df_sisagua['dia'] = df_sisagua['data_da_coleta'].dt.day # Extraindo o dia

df_sisagua['hora_da_coleta'] = pd.to_datetime(df_sisagua['hora_da_coleta'].str[:5] + ':00'
                                              , format='%H:%M:%S'
                                              , errors='coerce').dt.time # Convertendo o tipo do dado para datetime
```

```python
# Função para classificar o período do dia
def classificar_periodo(hora):
    if not isinstance(hora, time):
        return 'Horário inválido'

    hora_int = hora.hour # Para extrair apenas a hora

    if 5 <= hora_int < 12:
        return 'Manhã' # 05:00 às 11:59
    elif 12 <= hora_int < 18:
        return 'Tarde' # 12:00 às 17:59
    else:
        return 'Noite' # 18:00 às 23:59 e 00:00 às 04:59

# Aplicando a classificação
df_sisagua['periodo_do_dia'] = df_sisagua['hora_da_coleta'].apply(classificar_periodo)
```

```python
# Função para classificar o tipo da análise
def classificar_tipo_analise(parametro):

    parametro = str(parametro).strip().upper()
    if parametro in ['COLIFORMES TOTAIS', 'ESCHERICHIA COLI']:
        return 'Microbiológica'
    elif parametro in ['TURBIDEZ (UT)', 'CLORO RESIDUAL LIVRE (MG/L)', 'PH', 'COR APARENTE (UH)', 'FLUORETO (MG/L)']:
        return 'Físico-Química'
    return 'Outros'

# Aplicando a classificação
df_sisagua['tipo_analise'] = df_sisagua['parametro'].apply(classificar_tipo_analise)
```

```python
df_saa_micro = df_sisagua[(df_sisagua['tipo_da_forma_de_abastecimento'] == 'SAA') &
                   (df_sisagua['tipo_analise'] == 'Microbiológica')]
```

```python
# Função para converter resultado em binário
def resultado_bin(resultado):

    resultado = str(resultado).strip().upper()
    if resultado in ['PRESENTE']:
        return 1
    elif resultado in ['AUSENTE']:
        return 0
    return 'INVÁLIDO'
```
